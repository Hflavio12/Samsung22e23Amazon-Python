```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

## Recensioni del prodotto Samsumg Galaxy S22 ##

```python
# Creazione del driver di Selenium
driver = webdriver.Chrome()
```

```python
# Accesso alla pagina di accesso di Amazon
url = "https://www.amazon.it/gp/sign-in.html"
driver.get(url)
```

```python
url = "https://www.amazon.it/Caricatore-Cellulare-Smartphone-Fotocamere-Versione/product-reviews/B09R1VGP9D/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
driver.get(url)
```

```python
# Utilizza BeautifulSoup per trovare tutte le recensioni
all_reviews = []
while True:
    soup = BeautifulSoup(driver.page_source, "html.parser")
    reviews = soup.find_all("span", class_="review-text-content")
    all_reviews += [review.get_text() for review in reviews]

    # Cerca il pulsante "Successivo" e fai clic su di esso se √® disponibile
    next_buttons = driver.find_elements(By.CSS_SELECTOR, ".a-last > a")
    if next_buttons and "a-disabled" not in next_buttons[-1].get_attribute("class"):
        next_buttons[-1].click()
        time.sleep(3)
    else:
        break

# Stampa tutte le recensioni
for review in all_reviews:
    print(review)
```

```python
# Rimuovi tutte le stopwords dalle recensioni
stop_words = set(stopwords.words('italian'))
filtred_reviews = []
for review in all_reviews:
    words = review.lower().split()
    filtered_words = [word for word in words if word not in stop_words]
    filtred_reviews.append(' '.join(filtered_words))

# Stampa tutte le recensioni ad una stelle senza stopwords
for review in filtred_reviews:
    print(review)
    print()
```

#### Tag cloud sulle recensioni a 4 stelle di Samsumg S22 ####

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def create_tag_cloud(recensioni):
    for recensione in recensioni:
        if recensione['rating'] >= 4:
            # Rimuovi caratteri speciali e converti tutto in minuscolo
            plot_text = recensione['plot'].lower().replace('.', '').replace(',', '')

            # Dividi il testo in parole
            words = plot_text.split()

            # Conta le parole
            word_counts = Counter(words)

            # Seleziona le 50 parole pi√π comuni
            top_words = word_counts.most_common(50)

            # Crea la nuvola di tag
            wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(top_words))

            # Mostra la nuvola di tag
            plt.figure(figsize=(10, 5))
            plt.imshow(wordcloud, interpolation='bilinear')
            plt.title(recensione['title'])
            plt.axis('off')
            plt.show()

# Esempio di dati delle recensioni
recensioni = [
    {
        'title': 'Recensioni a 4 stelle di Samsung S22 (senza stopwords)',
        'rating': 4,
        'plot': 'ottimo smartphone, per√≤ va bene usa giorno....telefono perfetto tutto, nuovo ancora qualche bug niente che, pacco arriva scatola originale caricabatterie nero super velocebuonobisogno qualche aggiornamento software qualche problema, complesso ottimo smartphone '
    },
]

# Crea la tag cloud per le recensioni con rating = 4
create_tag_cloud(recensioni)
```
